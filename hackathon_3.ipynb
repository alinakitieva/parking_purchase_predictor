# Модель склонности клиента к приобретению машиноместа

Ключевую роль в продажах играет эффективная целевая рассылка. Рассылки позволяют оперативно информировать клиентов об актуальных предложениях и сервисах компании. Однако каждая рассылка сопряжена с различными издержками, что может негативно сказаться на лояльности клиентов в долгосрочной перспективе. На основе больших данных о предыдущем опыте взаимодействия с клиентами участникам хакатона предстоит разработать модель, позволяющую прогнозировать вероятность покупки клиентами дополнительных услуг, в частности, приобретения машиномест в паркинге. Разработанное решение позволит компании снизить затраты и улучшить лояльность клиентов за счет более персонифицированного подхода.

## Цель проекта

Разработать предсказательную модель на основе имеющихся данных о клиентах, позволяющую определить вероятность покупки машиноместа в паркинге в течение следующих трех месяцев. Это поможет компании оптимизировать целевую рассылку, сократить издержки и улучшить лояльность клиентов.

```python
# Визуализируем пропущенные значения
def plot_missing_data(df, interval=10):
    missing_data = df.isnull().sum() / len(df) * 100
    missing_data = missing_data[missing_data > 0].sort_values(ascending=False)

    n = len(missing_data.index)
    indices_to_show = range(0, n, interval)

    plt.figure(figsize=(10,6))
    sns.barplot(x=missing_data.index, y=missing_data.values)
    plt.xticks(rotation=90)
    plt.gca().set_xticks([], minor=False)
    plt.gca().set_xticks(indices_to_show, minor=False)
    plt.gca().set_xticklabels([missing_data.index[i] for i in indices_to_show], minor=False)
    plt.ylabel('Percentage of missing values (%)')
    plt.title('Percentage of missing data by feature')
    plt.show()
```

### Drop similar columns

```python
# Threshold for considering columns as similar
mse_threshold = 0.001  # Adjust this based on your needs

similar_cols = set()

for i, col1 in enumerate(df_new.select_dtypes(include=['int', 'float']).columns):
    for j, col2 in enumerate(df_new.select_dtypes(include=['int', 'float']).columns):
        if i < j:  # This ensures we're not comparing a column with itself and avoids redundant comparisons
            mse = mean_squared_error(df_new[col1].fillna(0), df_new[col2].fillna(0))
            if mse < mse_threshold:
                similar_cols.add(col2)

print(len(similar_cols), "columns were similar.")
```

```python
# Drop the similar columns
df_new.drop(columns=similar_cols, inplace=True)
```

### Redundancy check

```python
redundant_cols = set()
for i, col1 in enumerate(df_new.columns):
    for j, col2 in enumerate(df_new.columns):
        if i < j:
            if df_new[col1].equals(df_new[col2]):
                redundant_cols.add(col2)
len(redundant_cols)
```

### Выделение целевого признака

```python
df_new_sorted = df_new.sort_values(by="report_date").drop('report_date', axis=1)
```

```python
param_dist = {
    'n_estimators': [50, 100, 150, 200],
    'learning_rate': [0.01, 0.05, 0.1, 0.3, 0.5],
    'max_depth': [3, 4, 5, 6, 7, 8],
    'colsample_bytree': [0.5, 0.7, 0.9, 1],
    'subsample': [0.5, 0.7, 0.9, 1],
    'gamma': [0, 0.1, 0.2, 0.3, 0.4]
}

clf = xgb.XGBClassifier(objective='binary:logistic')
```

```python
tscv = TimeSeriesSplit(n_splits=5)

random_search = RandomizedSearchCV(
    clf,
    param_distributions=param_dist,
    n_iter=50,  # Number of parameter settings that are sampled (can be adjusted)
    scoring='roc_auc',
    n_jobs=-1,  # Use all available cores
    cv=tscv,
    verbose=3,
    random_state=42  # for reproducibility
)
```

```python
X = df_new_sorted.drop("target", axis=1)
y = df_new_sorted["target"]
```

```python
# Fitting RandomizedSearchCV
random_search.fit(X, y)

# Best parameters and model
best_params = random_search.best_params_
best_model = random_search.best_estimator_

print("Best parameters:", best_params)
```

Fitting 5 folds for each of 50 candidates, totalling 250 fits
Best parameters: {'subsample': 0.7, 'n_estimators': 150, 'max_depth': 7, 'learning_rate': 0.05, 'gamma': 0.1, 'colsample_bytree': 1}

```python
best_model = xgb.XGBClassifier(objective='binary:logistic', subsample='0.7', n_estimators=150, max_depth=7,
                              learning_rate=0.05,gamma=0.1,colsample_bytree=1)
```

```python
roc_auc_scores = []

for train_index, test_index in tscv.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    y_pred_prob = best_model.predict_proba(X_test)[:, 1]
    auc_score = roc_auc_score(y_test, y_pred_prob)
    
    roc_auc_scores.append(auc_score)

print("ROC AUC scores for the folds:", roc_auc_scores)
print("Average ROC AUC score:", sum(roc_auc_scores) / len(roc_auc_scores))
```
